```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding, Masking
from keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
import numpy as np

# Define the dataset
X_train = [[1, 2, 3], [4, 5], [6, 7, 8, 7]]
y_train = [0, 1, 0]

# Convert the labels to one-hot encoding
y_train = to_categorical(y_train)

# Find the maximum index value in the input data
max_index = max([max(seq) for seq in X_train])

# Pad the sequences with zeros so that they have the same length
X_train = pad_sequences(X_train, padding='post')

# Define the model architecture
model = Sequential()
model.add(Masking(mask_value=0, input_shape=(None,)))
model.add(Embedding(input_dim=max_index+1, output_dim=32))
model.add(LSTM(64))
model.add(Dense(2, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)

# Define the new data
X_test = [[9]]

# Pad the sequences with zeros
X_test = pad_sequences(X_test, padding='post', maxlen=X_train.shape[1])

# Make predictions on the new data
y_pred = model.predict(X_test)

# Convert the predicted probabilities to class labels
y_pred = np.argmax(y_pred, axis=1)

print(y_pred)
```
